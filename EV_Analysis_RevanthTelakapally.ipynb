```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')
```

```python
print ("Dataset Shape:",df.shape)
```

```python
print("\nColumns:\n")
print(df.columns)
print("\nDataset Info:\n")
df.info()
```

missing values


```python
missing_values= df.isnull().sum()
missing_values[missing_values >0]
```

DUPLICATES CHECK

```python
duplicates = df.duplicated().sum()
print("duplicate Rows:", duplicates)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("New Shape:", df.shape)
```

Handling Missing values

```python
df['Electric Range'] = df['Electric Range'].fillna(df['Electric Range'].median())
```

```python
df['Legislative District'] = df['Legislative District'].fillna(df['Legislative District'].median())
```

```python
df['County'] = df['County'].fillna(df['County'].mode()[0])

df['City'] = df['City'].fillna(df['City'].mode()[0])

df['Electric Utility'] = df['Electric Utility'].fillna(df['Electric Utility'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Postal Code'] = df['Postal Code'].fillna(df['Postal Code'].mode()[0])
```

```python
df['Vehicle Location'] = df['Vehicle Location'].fillna('Unckown Location')
```

```python
df.isnull().sum()
```

```python
df['2020 Census Tract'] = df['2020 Census Tract'].fillna(df['2020 Census Tract'].median())
```

```python
df.isnull().sum()
```

datatype check

```python
df.dtypes
```

```python
df['Model Year'] = df['Model Year'].astype(int)
```

```python
df['Postal Code'] = df['Postal Code'].astype(str)
print(df.dtypes)
```

```python
df.describe()
```

unique values check

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"\n{col}: {df[col].nunique()} unique values")
```

visualization

top 10 ev manufacturers

```python
top_make = df['Make'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_make.values, y=top_make.index)
plt.title('Top 10 Electric Vehicle Manufacturers')
plt.xlabel('Count')
plt.ylabel('Manufacturer')

plt.show()
```

EV Type distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='Electric Vehicle Type')

plt.title('Electric Vehicle Type Distribution')
plt.xticks(rotation=15)

plt.show()
```

Vehicle count by model year

```python
plt.figure(figsize=(14,6))

df['Model Year'].value_counts().sort_index().plot(kind='line', marker='o')

plt.title('Electric Vehicles by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Count')

plt.show()
```

TOP 10 CITIES WITH EV

```python
top_cities = df['City'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=top_cities.values, y=top_cities.index)

plt.title('Top 10 Cities with Electric Vehicles')
plt.xlabel('Count')
plt.ylabel('City')

plt.show()
```

Electric range distribution

```python
plt.figure(figsize=(10,5))

sns.histplot(df['Electric Range'], bins=30, kde=True)

plt.title('Electric Range Distribution')

plt.show()
```

CAFV eligibility count

```python
plt.figure(figsize=(12,5))

sns.countplot(data=df, y='Clean Alternative Fuel Vehicle (CAFV) Eligibility')

plt.title('CAFV Eligibility Distribution')

plt.show()
plt.figure(figsize=(12,5))

sns.countplot(data=df, y='Clean Alternative Fuel Vehicle (CAFV) Eligibility')

plt.title('CAFV Eligibility Distribution')

plt.show()
```

Top 10 EV models

```python
top_models = df['Model'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(x=top_models.values, y=top_models.index)

plt.title('Top 10 EV Models')
plt.xlabel('Count')
plt.ylabel('Model')

plt.show()
```

Correlation heatmap

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,5))

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title('Correlation Heatmap')

plt.show()
```

STATE WISE VEHICLE COUNT

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df, x='State', order=df['State'].value_counts().index[:10])

plt.title('State-wise Electric Vehicle Count')

plt.show()
```

BOXPLOT FOR ELECTRIC RANGE

```python
plt.figure(figsize=(10,5))

sns.boxplot(x=df['Electric Range'])

plt.title('Boxplot of Electric Range')

plt.show()
```
